#Análise do domínio

Neste notebook, vou explorar a origem e formato dos dados que temos para experimentar.

##Origem dos dados

Estes dados são originados de experimentos farmaceuticos realizados com milhares de compostos e repetidos dezenas ou centenas de vezes para obtermos resultados estatisticamente significantes.

Esta análise exploratória só é possível em larga escala com a automatização dos laboratórios:

Para o nosso caso, temos três principais medições que são realizadas em cada experimento, que são coletados de forma automatizada ou manual:

1.   Taxa de expressão genética *(automatizada)*
2.   Taxa de sobrevivência de células *(automatizada)*
3.   Mecanismos de Ação (MoA) *(manual)*



###Taxa de expressão genética
Nos informa a taxa de produção de **RNA** de um gene específico, dadas as condições do experimento. Isto dá uma informação ao pesquisador de quais são os genes mais ativos ou inativos em resposta às condições do experimento

###Taxa de sobreviência de células
A taxa de sobrevivência de alguns tipos celulares são analisadas quando expostas às condições do experimento.

Segue a imagem de um mecanismo para análise das células individualmente

![mecanismo](https://raw.githubusercontent.com/LPugens/imersao-dados-desafio-final/main/fig2_mecanismo_celulas.jpg)

Fonte: [artigo da Science](https://www.nature.com/articles/nrd.2017.232.pdf?origin=ppub)

###Mecanismos de Ação (MoA)
São efeitos encontrados nos experimentos realizados. Estes efeitos podem ser desde a inibição de patógenos até efeitos diuréticos e produção de vitaminas encontrados analisando as amostras com microscopia manualmente.

##Formato dos dados

Vamos extraír algumas informações dos dados

###Dados de experimentos
Aqui temos os dados extraídos de maneira automatizada pelos laboratórios.

```python
dados_experimentos = 'https://github.com/LPugens/imersao-dados-desafio-final/blob/main/Dados/dados_experimentos.zip?raw=true'
dados_experimentos = pd.read_csv(dados_experimentos, compression = 'zip')
dados_experimentos.head()
```

Analisando as colunas do nosso dataset, temos primeiramente o `id` do experimento, que é um identificador da cultura que foi experimentada. 

Em seguida temos algumas colunas que identificam as características do experimento:
*   `tratamento`: Identifica experimentos de controle (sem droga) e experimentos nos quais foram introduzidas drogas
*   `tempo`: Por quanto tempo estes experimentos foram expostos às condições
*   `dose`: Volume da droga administrado (assumo que `D1` < `D2`)
*   `droga`: Identificador anonimizado de qual droga foi administrada no experimento

Logo em seguida, temos dois conjuntos de colunas identificando **expressões genéticas** e **taxa de sobrevivência celular**


###Dados de resultados
Aqui temos os dados extraídos de maneira manual pelos pesquisadores ao analisar os experimentos conduzidos de maneira automatizada.

Estes dados são valiosos, uma vez que é necessário um profissional especializado e muito tempo para extraílos.

Também são os dados mais relevantes, pois indicam os efeitos encontrados com as condições impostas pelo experimento

```python
dados_resultados = 'https://github.com/LPugens/imersao-dados-desafio-final/blob/main/Dados/dados_resultados.csv?raw=true'
dados_resultados = pd.read_csv(dados_resultados)
dados_resultados.head()
```
